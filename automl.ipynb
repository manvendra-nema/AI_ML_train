```python
import pandas as pd
from pycaret.classification import setup, compare_models, save_model, predict_model

# Step 1: Read the saved CSV files
train_data = pd.read_csv(r'D:/Downloads/dataset/X_train_cleaned.csv')
test_data = pd.read_csv(r'D:/Downloads/dataset/X_test_cleaned.csv')

# Read the target variable (assuming sentiment is the target)
Y_train = train_data['sentiment']
Y_test = test_data['sentiment']

# Read the feature variable (assuming review_body is the feature)
X_train_cleaned = train_data['review_body']
X_test_cleaned = test_data['review_body']

# Combine X and Y into a single DataFrame for PyCaret
train_data_pycaret = pd.DataFrame({'review_body': X_train_cleaned, 'sentiment': Y_train})
test_data_pycaret = pd.DataFrame({'review_body': X_test_cleaned, 'sentiment': Y_test})

# Step 2: Initialize PyCaret's setup for classification (training)
exp = setup(data=train_data_pycaret, 
            target='sentiment', 
            train_size=0.8, 
            test_data=test_data_pycaret, 
            index=False,
            fold=5)
```

```python
# Step 3: Compare models and select the best one
best_model = compare_models()

# Step 4: Save the best model for deployment
save_model(best_model, "best_automl_model")

# Step 5: Evaluate the best model on the test data
results = predict_model(best_model, data=test_data_pycaret)
print("Predictions on Test Data:")
print(results.head())

# Optionally, you can load the saved model later to predict new data
# loaded_model = load_model("best_automl_model")
# predictions = predict_model(loaded_model, data=new_data)
```
